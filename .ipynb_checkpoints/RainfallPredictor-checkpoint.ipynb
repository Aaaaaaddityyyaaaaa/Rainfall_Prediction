```python
def map(date):
    month = date.month
    if(month==3 or month==4 or month==5):
        return 'summer'
    elif(month==6 or month==7 or month==8):
        return 'autumn'
    elif(month==9 or month==10 or month==11):
        return 'winter'
    else :
        return 'spring'
```

```python
df["Date"]  = pd.to_datetime(df.Date)
```

```python
df["Season"] = df["Date"].apply(map)
```
