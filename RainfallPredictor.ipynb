```python
def map(date):
    month = date.month
    if(month==3 or month==4 or month==5):
        return 'summer'
    elif(month==6 or month==7 or month==8):
        return 'autumn'
    elif(month==9 or month==10 or month==11):
        return 'winter'
    else :
        return 'spring'
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler , OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split , GridSearchCV , StratifiedKFold
```

```python
x.dtypes
```

```python
categorical_cols = x.select_dtypes(include="object") 
```

```python
numerical_cols = x.select_dtypes(include=["float64"])
```

```python
df.drop("Date" , axis=1 , inplace = True)
```

```python
df
```

```python
x.drop("Date" ,axis=1 , inplace=True)
```

```python
numerical_cols = numerical_cols.columns.to_list()
```

```python
categorical_cols = categorical_cols.columns.to_list()
```

```python
# Scale the numeric features
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

# One-hot encode the categoricals 
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ]
)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline = Pipeline(steps=[('preprocess'  , preprocessor),('train' , RandomForestClassifier(random_state=42))])
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y,test_size=0.2 , stratify=y,random_state=42)
```

```python
param_grid = {
    'train__n_estimators': [50, 100],
    'train__max_depth': [None, 10, 20],
    'train__min_samples_split': [2, 5]
}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
gridsearch = GridSearchCV(pipeline,param_grid,cv=cv,scoring="accuracy" , verbose=2)
```

```python
gridsearch.fit(xtrain,ytrain)
```

```python
bestpipeline = gridsearch.best_estimator_
```

```python
ypred = bestpipeline.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy  = accuracy_score(ytest , ypred)
```
